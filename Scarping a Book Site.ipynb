### here the problem lies with rating list and availability list

#### in rating list , the element itself is a list but we need a single value, that's why we would be using .attrs["class"][1] 
#### and in availability list , we would strip all the white spaces (new line , space , tab)


```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

html = urlopen("http://books.toscrape.com")
bsObj = BeautifulSoup(html.read())

books = bsObj.findAll("article",{"class":"product_pod"})

title = list()
rating = list()
price = list()
availability = list()

for book in books:
    title.append(book.find("h3").find("a").attrs["title"])
    rating.append(book.find("p").attrs["class"][1])
    price.append(book.find("div",{"class":"product_price"}).find("p",{"class":"price_color"}).get_text())
    availability.append(book.find("div",{"class":"product_price"}).find("p",{"class":"instock availability"}).get_text().strip())

print(title)
print(rating)
print(price)
print(availability)
```

# Making Panda DataFrame

## Scraping only first page

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

html = urlopen("http://books.toscrape.com")
bsObj = BeautifulSoup(html.read())

books = bsObj.findAll("article",{"class":"product_pod"})

title = list()
rating = list()
price = list()
availability = list()

for book in books:
    title.append(book.find("h3").find("a").attrs["title"])
    rating.append(book.find("p").attrs["class"][1])
    price.append(book.find("div",{"class":"product_price"}).find("p",{"class":"price_color"}).get_text())
    availability.append(book.find("div",{"class":"product_price"}).find("p",{"class":"instock availability"}).get_text().strip())

df = pd.DataFrame({
    'Title' : title,
    'Star Rating' : rating,
    'Price' : price,
    'Availability' : availability
}, columns = ['Title','Star Rating','Price','Availability'])    

df
```

## Scraping Multiple Page

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

url = "http://books.toscrape.com"
html = urlopen(url)
bsObj = BeautifulSoup(html.read())

page_no = int(input('From how many number of pages do you want to scarp? : '))

title = list()
rating = list()
price = list()
availability = list()
page = list()

i = 1
while i<=page_no:
    
    books = bsObj.findAll("article",{"class":"product_pod"})

    for book in books:
        title.append(book.find("h3").find("a").attrs["title"])
        rating.append(book.find("p").attrs["class"][1])
        price.append(book.find("div",{"class":"product_price"}).find("p",{"class":"price_color"}).get_text())
        availability.append(book.find("div",{"class":"product_price"}).find("p",{"class":"instock availability"}).get_text().strip())
        page.append(i)
        
    i = i + 1
    next_url = bsObj.find("li",{"class":"next"}).find("a").attrs['href']
    if "catalogue/" not in next_url:
        next_url = "catalogue/" + next_url
    print(url+"/"+next_url)
    html = urlopen(url+"/"+next_url)
    bsObj = BeautifulSoup(html.read())

df = pd.DataFrame({
    'Title' : title,
    'Star Rating' : rating,
    'Price' : price,
    'Availability' : availability,
    'Page No' : page
}, columns = ['Title','Star Rating','Price','Availability','Page No'])    

df
```
